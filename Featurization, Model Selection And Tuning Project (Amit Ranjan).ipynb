```python
#using PCA dimensions reduction technique to reduce the number of unwanted columns from the dataset.
#importing the necessary libraries.

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
#removing the target variable from the dataset to scale & normalize the dataset.

y = Data1.pop('Pass/Fail')
```

```python
scalar = StandardScaler()
df_scaled = pd.DataFrame(scalar.fit_transform(Data1), columns=Data1.columns)
df_scaled
```

```python
df_scaled.reset_index(drop=True)
```

```python
pca = PCA()
df_pca = pd.DataFrame(pca.fit_transform(df_scaled))
df_pca.shape
```

```python
explained_variance = np.cumsum(pca.explained_variance_ratio_)
plt.figure(1,figsize=(15,8))
plt.ylim(0.0,1.1)

plt.clf()

plt.plot(list(range(1,443)), explained_variance, marker='o', linestyle='--', color='green')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance")

plt.title("The number of components needed to explain variance")

plt.axhline(y=0.90, color='r', linestyle='--')
plt.axhline(y=0.93, color='b', linestyle='--')

plt.axvline(x=140, color='r', linestyle='--')
plt.grid(True)
plt.show()
```

```python
#Observation:From the above plot we can say that selecting minimum component=140,the cumulative variance shows at 90% mark.
# as the elbow method points to selecting the number of components equal to 140 will be the best fit for this dataset.
```

```python
pca_new = PCA(n_components=140)
pca_new.fit(df_pca)

npca_new = pca_new.transform(df_pca)
#Here, the npca_new is our training set for further designing the model.
print(pca_new.explained_variance_ratio_)
```

```python
#PART 2D
#checking for multi-collinearity in the data and taking the appropriate action on the dataset.

#Detecting the multi-collinearity between the variables using the Variance Inflation Factor(VIF) technique.
#imporing the necessary libraries.

from statsmodels.stats.outliers_influence import variance_inflation_factor

#initializing the VIF dataframe

n = pd.DataFrame(npca_new)

vif_data = pd.DataFrame()
vif_data["feature"] = n.columns
vif_data["VIF"] = [variance_inflation_factor(n.values, i) for i in range(len(n.columns))]
print(vif_data)
```

```python
#PART 5B
#using the cross validation technique KFold
#importing the necessary technique.

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

num_folds = 50
seed = 10

X = XScaled
y = y

X_train, X_test, y_train, y_test = train_test_split(XScaled, y, test_size=0.50, random_state=1)
kfold = KFold(n_splits=num_folds, random_state=seed, shuffle=True)

model = LogisticRegression()
results = cross_val_score(model, X, y, cv=kfold)
print(results)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100))
```

```python
#Observation: The avaerage of all the 50 accuracy scores is coming out as 92.016% with
#             standard deviation of 4.327% on either sides.
```

```python
#PART 5C
#Applying the hyper-parameter technique RandomizedSearchCV & GradientBoostingRegressor to get the best accuracy.

#importing necessary libraries.

from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from scipy.stats import uniform as sp_randFloat
from scipy.stats import randint as sp_randInt

X_train, X_test, y_train, y_test = train_test_split(XScaled, y, test_size=0.25, random_state=1)
model = GradientBoostingRegressor()

parameters = {'learning_rate': sp_randFloat(),
                  'subsample'    : sp_randFloat(),
                  'n_estimators' : sp_randInt(100,1200),
                  'max_depth'    : sp_randInt(4,10)
                 }

randm_src = RandomizedSearchCV(estimator=model, param_distributions = parameters,
                               cv = 3, n_jobs=-1, n_iter = 7)

randm_src.fit(X_train, y_train)

print(" Results from Random Search " )
print("\n The best estimator across ALL searched params:\n", randm_src.best_estimator_)

print("\n The best score across ALL searched params:\n", abs(randm_src.best_score_))
print("\n The best parameters across ALL searched params:\n", randm_src.best_params_)
```

```python
#Observation: Here, we can see that the n_estimators=1077 is selected whereas
# learning_rate of 0.171923964838668 are selected. Similarly, max_depth of 8 is selected as well.

#Now finally, we will build our final classifier model by passing the above parameters.

X_train, X_test, y_train, y_test = train_test_split(XScaled, y, test_size=0.25, random_state=1)

clf2 = GradientBoostingRegressor(learning_rate=0.171923964838668, max_depth=8, n_estimators=1077,
                                 subsample=0.3971302115116522)

clf2.fit(X_train, y_train)
prediction=clf2.predict(X_test)
print("Train score: ",clf2.score(X_train, y_train))
print("Test score: ",abs(clf2.score(X_test, y_test)))
```

```python
#Observtion: After hyper-parameter tuning using RandomizedSearchCV we are getting 84.68% train score & 45.86% test score.
```

```python
#PART 5D
#further enchancing the model performance, here using the target variable balancing method using SMOTE.

#importing the necessary library.

from imblearn.over_sampling import SMOTE

print("Before UpSampling, counts of label '-1.0': {}".format(sum(y_train==-1)))
print("Before UpSampling, counts of label '1.0': {} \n".format(sum(y_train==1)))
```

```python
sm = SMOTE(sampling_strategy = 1 ,k_neighbors = 5, random_state=1)   #Synthetic Minority Over Sampling Technique
X_train_res, y_train_res = sm.fit_resample(X_train, y_train.ravel())

print("After UpSampling, counts of label '-1.0': {}".format(sum(y_train_res==-1)))
print("After UpSampling, counts of label '1.0': {} \n".format(sum(y_train_res==1)))
```

```python
print('After UpSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After UpSampling, the shape of train_y: {} \n'.format(y_train_res.shape))
```

```python
# fit model on upsampled data 

logistic_model.fit(X_train_res, y_train_res)
y_predict = logistic_model.predict(X_test)
upsampled_model_score = logistic_model.score(X_test, y_test)
print(upsampled_model_score)
print(metrics.confusion_matrix(y_test, y_predict))
print(metrics.classification_report(y_test, y_predict))
```
